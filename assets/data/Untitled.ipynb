```python
import pandas as pd
```

```python
projects = pd.read_csv("./extracted_project_data.csv")
p_s = pd.read_csv("./project_subject_area.csv")
subject = pd.read_csv("./subject_areas.csv")
```

```python
p_s
```

```python
projects = projects.merge(p_s, left_on="project_id_number", right_on="project_id")
projects = projects.merge(subject, left_on="subject_area", right_on="subject_area")
```

```python
projects["research_area"] = projects["research_area"].apply(lambda x : x[:x.find("(")])
```

```python
projects[["project_abstract", "project_id", "subject_area", "review_board", "research_area"]][:100].to_csv("./train_small.csv")
```
